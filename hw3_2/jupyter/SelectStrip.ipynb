# 参考条选择


```c++
TFile *ipf = new TFile("../data/data_16Cs.root", "read");
TTree *tree = (TTree*)ipf->Get("tree");
TCanvas *c1 = new TCanvas;
```

## dssd1

寻找xy两面关联最多的作为参考

```c++
%jsroot on
tree->Draw("d1xs>>h1xs(32, 0, 32)", "", "");
tree->Draw("d1ys>>h1ys(32, 0, 32)", "", "same");
TH1 *h1xs = (TH1*)gDirectory->Get("h1xs");
h1xs->SetLineColor(kRed);
c1->Draw();
```

```c++
tree->Draw("d1xhit:d1yhit>>(6, 0, 6, 6, 0, 6)", "", "colz");
c1->Draw();
```

```c++
TCut d1xhit1Cut = "d1xhit == 1";
TCut d1yhit1Cut = "d1yhit == 1";
TCut d1hit1Cut = d1xhit1Cut && d1yhit1Cut;
TCut d1xhit2Cut = "d1xhit == 2 && abs(d1xs[0]-d1xs[1]) > 1";
TCut d1yhit2Cut = "d1yhit == 2 && abs(d1ys[0]-d1ys[1]) > 1";
TCut d1hit2Cut = d1xhit2Cut && d1yhit2Cut;
```

因为已经将能量降序排列，所以直接画出`d1ys:d1xs`而不用考虑下标。

```c++
tree->Draw("d1ys:d1xs>>h2d1xy11(32, 0, 32, 32, 0, 32)", d1hit1Cut, "colz");
c1->Draw();
```

```c++
tree->Draw("d1ys:d1xs>>h2d1xy22(32, 0, 32, 32, 0, 32)", d1hit2Cut, "colz");
c1->Draw();
```

```c++
TH2* h2d1xy11 = (TH2*)gDirectory->Get("h2d1xy11");
TH2* h2d1xy22 = (TH2*)gDirectory->Get("h2d1xy22");
h2d1xy22->Add(h2d1xy22, h2d1xy11);
h2d1xy22->Draw("colz");
c1->Draw();
```

由此应该选择x面第17根条作为参考条，对y面第9到16条进行归一化

```c++
tree->Draw("d1xs>>hd1n11(32, 0, 32)", d1hit1Cut && "d1ys <= 15 && d1ys >= 8");
c1->Draw();
```

```c++
tree->Draw("d1xs>>hd1n12(32, 0, 32)", d1hit2Cut && "d1ys <= 15 && d1ys >= 8");
c1->Draw();
```

```c++
TH1 *hd1n11 = (TH1*)gDirectory->Get("hd1n11");
TH1 *hd1n12 = (TH1*)gDirectory->Get("hd1n12");
hd1n12->Add(hd1n11, hd1n12);
hd1n12->Draw();
c1->Draw();
```

接着选择x面第12到第24根条

```c++
tree->Draw("d1ys>>hd1n21(32, 0, 32)", d1hit1Cut && "d1xs <= 23 && d1xs >= 11");
c1->Draw();
```

```c++
tree->Draw("d1ys>>hd1n22(32, 0, 32)", d1hit2Cut && "d1xs <= 23 && d1xs >= 11");
c1->Draw();
```

```c++
TH1 *hd1n21 = (TH1*)gDirectory->Get("hd1n21");
TH1 *hd1n22 = (TH1*)gDirectory->Get("hd1n22");
hd1n21->Add(hd1n21, hd1n22);
hd1n21->Draw();
c1->Draw();
```

接着选择y面所有条，x面所有条。

总体流程是 
**`x[16..17) --> y[9..16) --> x[11..24) --> y[0..32) --> x[0..32)`**
